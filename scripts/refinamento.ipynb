```python
import duckdb
```

```python
con = duckdb.connect(database='dados_duckdb.db', read_only=False)
```

```python
df = con.execute("SELECT * FROM produtos").fetchdf()
df.head()
```

```python
# criar dimens√£o na camada gold, removendo algumas colunas
df2 = df.drop(columns=['id_categoria', 'id_fornecedor'])
df2.head(10)
```

```python
# Criar tabela na camada gold
con.execute("""
    CREATE TABLE IF NOT EXISTS dim_produtos(
            id_produto BIGINT,
            nm_produto TEXT,
            vl_preco FLOAT
            )
""")
```

```python
df2.head()
```

```python
# Inserir valores a tabela
con.execute("""INSERT INTO dim_produtos SELECT * FROM df2""")
```

```python
df_dim = con.execute("SELECT * FROM dim_produtos").fetchdf()
df_dim.head()
```
